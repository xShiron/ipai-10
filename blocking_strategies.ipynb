### Blocking

#### NBA dataset

```python
nba_blocking = nba.copy()
nba_blocking["block"] = nba_blocking["season"].astype(str) + nba_blocking["team1"] + nba_blocking["team2"]
nba_blocking
```

This blocking strategy effectively makes it possible to access any of the records by just using the season and abbreviations of the teams that played.

Blocking strategy: Match_Season + Team1 + Team2

#### US Traffic accidents dataset

```python
accidents_blocking = us_accidents.copy()
```

```python
accidents_blocking["block"] = accidents_blocking["Start_Time"].str.split("-").apply(lambda x: x[0]) + accidents_blocking["City"] + accidents_blocking["State"]
accidents_blocking["block"]
```

This blocking strategy makes it possible to access records faster through the year, state and city where they occurred.

Blocking strategy: Year + City + State

#### Air Pollution dataset
